```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
var=sns.load_dataset('tips')
var
```

```python
var.isnull().values.any()
```

```python
sns.heatmap(var.corr(),annot=True)
```

```python
sns.boxplot(x='time',y='total_bill',data=var)
```

**Most of the tips come from the dinner**

```python
sns.boxplot(x='time',y='total_bill',data=var,hue='sex')
```

**Females pay less tip than male in the dinner**

```python
sns.scatterplot(x='size',y='total_bill',data=var)
```

```python
sns.histplot(data=var,x='tip',kde=True,element='step')
```

Most of the tips are between 2 to 4 dollars

```python
sns.scatterplot(data=var,x='size',y='tip')
```

```python
sns.barplot(data=var,x='size',y='tip',ci=-0)
```

```python
sns.lineplot(data=var,x='size',y='tip',ci=-0)
```

**The tip increases as the food or order size increases**

```python
sns.kdeplot(data=var,x='tip',y='total_bill',hue='smoker',dodge=True)
```

```python
sns.boxplot(data=var,x='day',y='total_bill',hue='smoker')
```

```python
sns.barplot(data=var,x='day',y='total_bill',hue='smoker',errorbar=('ci', 0))
plt.show()
```

Most of the bills are paid by smokers on sunday

```python
smoker=var.smoker.value_counts().to_dict()
```
